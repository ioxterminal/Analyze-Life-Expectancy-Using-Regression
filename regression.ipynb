```python
#renaming column names
crime =crime.set_axis(['State','PopTot2012', 'MurderNMTot2015', 'MurderTot2015', 'GunMurderTot2015', 'GunOwnPer2013', 'MurderNMRate_per_100k_2015', 'MurderRate_per_100k_2015', 'GunMurderRate_per_100k_2015'], axis=1) #change the column names
```

```python
#the dataset contain non numeric values like "—[a]" and must be cleaned
print(crime) #print the full set
```

```python
#line 51 to 61 do not contain useful data an can be deleted
#the dataset contain non numeric values like "—[a]" "—[b]" and must be changed to NaN  

crime = crime.replace(to_replace ='.\s\[a\]', value = 'NaN', regex = True)   #— [a] with a blank will be changed to NaN
crime = crime.replace(to_replace ='.\s\[b\]', value = 'NaN', regex = True)   #— [a] without a blank will be changed to NaN
crime = crime.replace(to_replace ='.\[a\]', value = 'NaN', regex = True)     #— [b]  with a blank will be changed to NaN
crime = crime.replace(to_replace ='.\[b\]', value = 'NaN', regex = True)     #— [b]  without a blank will be changed to NaN
crime = crime.replace(to_replace =',', value = '', regex = True)

print(crime) #after replacement of values like "—[a]" "—[b]" with and without space we need to check the dataset
```

```python
#NaN values look fine   
crime.isna() # verify NaN values are detected as such
```

```python
#NaN values looks fine
crime = crime.dropna(thresh=2) #Keep only the rows with at least 2 non-NA values. This should delete lines 51-61 as well
```
